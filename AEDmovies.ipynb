```python
import pandas as pd
import numpy as np
```

```python
dados = pd.read_csv("https://raw.githubusercontent.com/WilliBobadilla/datasetmovies/master/movies.csv")
```

```python
#visualizando as linhas 
dados.head(76)
```

```python
dados.columns = ['film','genre','lead_studio','audience_score','Profitability','rotten_tomatoes','worldwide_gross','year']
```

```python
#Audience Score
dados.audience_score
```

# **Medidas Audience Score**

```python
# MÃ©dia:
media = np.mean(dados.audience_score)
print("MÃ©dia:", media)

# Moda,nÃ£o tem

# Mediana
mediana = np.median(dados.audience_score)
print("Mediana:", mediana)

# Amplitude
amplitude = np.max(dados.audience_score) - np.min(dados.audience_score)
print("Amplitude:", amplitude)

# Desvio mÃ©dio
desvio_medio = np.mean(np.abs(dados.audience_score - np.mean(dados.audience_score)))
print("Desvio mÃ©dio:", desvio_medio)

# Desvio mÃ©dio absoluto
desvio_medio_absoluto = np.median(np.abs(dados.audience_score - np.median(dados.audience_score)))
print("Desvio mÃ©dio absoluto:", desvio_medio_absoluto)

# VariÃ¢ncia
variancia = np.var(dados.audience_score, ddof=1)
print("VariÃ¢ncia:", variancia)

# Desvio padrÃ£o
desvio_padrao = np.std(dados.audience_score, ddof=1)
print("Desvio padrÃ£o:", desvio_padrao)
```

# **BoxPlot Audience Score**

```python
import numpy as np
dados= dados.audience_score

# Primeiro quartil (25%)
q1 = np.percentile(dados, 25)
print("Primeiro quartil:", q1)

# Segundo quartil (50%)
q2 = np.percentile(dados, 50)
print("Segundo quartil (mediana):", q2)

# Terceiro quartil (75%)
q3 = np.percentile(dados, 75)
print("Terceiro quartil:", q3)

# Amplitude interquartil (IQR)
iqr = q3 - q1
print("Amplitude interquartil (IQR):", iqr)

# Limite inferior
lim_inf = q1 - 1.5 * iqr
print("Limite inferior:", lim_inf)

# Limite superior
lim_sup = q3 + 1.5 * iqr
print("Limite superior:", lim_sup)

#para montar o boxplot
import matplotlib.pyplot as plt

plt.boxplot(dados)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Filtrar valores ausentes
dados = dados.dropna(subset=['audience_score'])

# Ordenar os dados pelo audience_score
dados = dados.sort_values(by='audience_score', ascending=False)

# Selecionar os 10 filmes com maior audience_score
top_10_filmes = dados.head(10)

# Criar o grÃ¡fico de barras
plt.bar(top_10_filmes['film'], top_10_filmes['audience_score'])

# Personalizar o grÃ¡fico
plt.title('Top 10 Filmes com Maior NÃºmero de AudiÃªncia')
plt.xlabel('Filme')
plt.ylabel('Audience Score')

# Rotacionar os rÃ³tulos do eixo x
plt.xticks(rotation=90)

# Exibir o grÃ¡fico
plt.show()
```
