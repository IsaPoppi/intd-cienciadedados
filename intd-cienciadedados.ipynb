# **Calculando em Python,agora sem importar banco de dados para analisar:**


```python
import numpy as np

dados = [10, 12, 15, 20, 25, 30, 35, 40, 45, 50]

# Média
media = np.mean(dados)
print("Média:", media)

# Moda,não tem

# Mediana
mediana = np.median(dados)
print("Mediana:", mediana)

# Amplitude
amplitude = np.max(dados) - np.min(dados)
print("Amplitude:", amplitude)

# Desvio médio
desvio_medio = np.mean(np.abs(dados - np.mean(dados)))
print("Desvio médio:", desvio_medio)

# Desvio médio absoluto
desvio_medio_absoluto = np.median(np.abs(dados - np.median(dados)))
print("Desvio médio absoluto:", desvio_medio_absoluto)

# Variância
variancia = np.var(dados, ddof=1)
print("Variância:", variancia)

# Desvio padrão
desvio_padrao = np.std(dados, ddof=1)
print("Desvio padrão:", desvio_padrao)

#Observe que para a variância e o desvio padrão, o argumento ddof=1foi usado 
#para calcular a variância e o desvio padrão amostral em vez da população inteira. 
#Se você quiser calcular a variância e o desvio padrão da população inteira,
#deixe o argumento ddofcom o valor padrão de 0.
```

Calculando o boxplot

```python
import numpy as np

dados = [10, 12, 15, 20, 25, 30, 35, 40, 45, 50]

# Primeiro quartil (25%)
q1 = np.percentile(dados, 25)
print("Primeiro quartil:", q1)

# Segundo quartil (50%)
q2 = np.percentile(dados, 50)
print("Segundo quartil (mediana):", q2)

# Terceiro quartil (75%)
q3 = np.percentile(dados, 75)
print("Terceiro quartil:", q3)

# Amplitude interquartil (IQR)
iqr = q3 - q1
print("Amplitude interquartil (IQR):", iqr)

# Limite inferior
lim_inf = q1 - 1.5 * iqr
print("Limite inferior:", lim_inf)

# Limite superior
lim_sup = q3 + 1.5 * iqr
print("Limite superior:", lim_sup)

#para montar o boxplot
import matplotlib.pyplot as plt

dados = [10, 12, 15, 20, 25, 30, 35, 40, 45, 50]

plt.boxplot(dados)
plt.show()
```

## **Boxplot**:
Para interpretá-lo antes precisamos entender cada um desses elementos:

Um boxplot é um tipo de gráfico usado para visualizar a distribuição de um conjunto de dados. Ele é composto por um retângulo (que representa os quartis) e por duas linhas que se estendem a partir do retângulo (que representam os valores mínimo e máximo dos dados).

Aqui está como interpretar um boxplot:

Mediana:

 A linha dentro do retângulo representa a mediana dos dados, ou seja, o valor que separa 50% dos dados maiores dos 50% menores.

Quartis:

 O retângulo do boxplot representa o intervalo interquartil, que é a distância entre o primeiro quartil (Q1) e o terceiro quartil (Q3). O primeiro quartil é o valor que separa os 25% menores dos dados, enquanto o terceiro quartil separa os 25% maiores dos dados.

Os quartis são medidas estatísticas que dividem um conjunto de dados ordenados em quatro partes iguais. Cada quartil representa um determinado percentual dos dados e pode fornecer informações importantes sobre a distribuição desses dados.

Os três quartis principais são o primeiro quartil (Q1), o segundo quartil (Q2) e o terceiro quartil (Q3), e eles dividem os dados em quatro partes iguais, cada uma contendo 25% dos dados. O primeiro quartil, Q1, representa o valor abaixo do qual se encontra 25% dos dados, enquanto o terceiro quartil, Q3, representa o valor abaixo do qual se encontra 75% dos dados. O segundo quartil, Q2, é simplesmente a mediana dos dados, que separa os dados em duas partes iguais, cada uma contendo 50% dos dados.

Em resumo, cada quartil representa uma especificidade específica dos dados e fornece informações importantes sobre a distribuição desses dados. O primeiro quartil, Q1, representa o valor abaixo do qual se encontra 25% dos dados, o segundo quartil, Q2, é a mediana, que divide os dados em duas partes iguais, e o terceiro quartil, Q3, representa o valor abaixo do qual se encontra 75% dos dados.


-Os limites superior e inferior

Além dos quartis, um boxplot é composto pelos limites superior e inferior. Esses limites correspondem, respectivamente, ao maior e menor valor do conjunto de dados que não sejam outliers. Um outlier é um valor que destoa do conjunto de dados, o que podemos chamar em português de “valor discrepante” - mas aqui, de novo, o termo em inglês é mais conhecido e usado.


E como definir se um valor é outlier?

 Para isso, precisamos calcular os limites superior e inferior teóricos. Esses limites são definidos com base na amplitude interquartil (AIQ). Da seguinte forma:

-Limite inferior teórico = 
Q1-1,5* AIQ

-Limite superior teórico = 
Q3+1,5* AIQ

# ***Definir conceitos:***

# **O que é uma variável independente?**
Uma variável independente é a variável que é alterada ou controlada em um experimento científico para testar os efeitos sobre a variável dependente.

# **O que é uma variável dependente?**
Uma variável dependente é a variável que está sendo testada e medida em um experimento científico.

# **Exemplo de Variável Livre e Dependente**
Por exemplo, um pesquisador precisa verificar se a intensidade da luz tem algum impacto sobre uma traça que está sendo puxada para dentro dela. A intensidade da luz é controlada pelo pesquisador. Esta seria a variável independente. Como a traça responde aos níveis de luz distintos (separação da fonte de luz) seria a variável dependente.

# **Instruções passo a passo para distinguir as variáveis**
-As variáveis independentes e dependentes são os dois fatores-chave em um experimento científico.
A variável independente é aquela que o experimentador controla. A variável dependente é a variável que se ajusta por causa da variável independente.
As duas variáveis podem estar conectadas por circunstâncias e resultados lógicos. Se a variável independente muda e elas estão relacionadas, então a variável dependente é influenciada.

# **Features** 
Em aprendizado de máquina, "features" (características) referem-se a variáveis ​​ou atributos que são usados ​​para descrever um objeto ou amostra de dados. As características são os dados de entrada que são usados ​​para treinar um modelo de aprendizado de máquina para realizar uma tarefa específica, como classificação ou previsão.

As características podem ser numéricas, categóricas, binárias ou textuais, dependendo do tipo de problema que se está tentando resolver. Por exemplo, em um problema de classificação de imagens, as características podem ser as dimensões de uma imagem, o número de pixels, as intensidades dos pixels em diferentes canais de cor, entre outros. Em um problema de análise de sentimentos em texto, as características podem incluir a frequência de palavras específicas em um conjunto de dados.

A escolha das características é uma etapa importante no processo de modelagem de aprendizado de máquina, pois a qualidade e fidelidade das características escolhidas têm um impacto significativo no desempenho do modelo. O processo de seleção de recursos geralmente envolve a exploração dos dados e a escolha dos recursos que são mais relevantes para o problema em questão.

As características são usadas para alimentar o modelo de aprendizado de máquina, que aprenderá a identificar padrões nos dados e usar esses padrões para fazer previsões ou tomar decisões. À medida que o modelo é treinado com mais dados, ele pode ser ajustado para melhorar seu desempenho na tarefa de interesse.

 ***-Classificação: ***

Os problemas de classificação são aqueles onde se busca encontrar uma classe, dentro das possibilidades limitadas existentes. 

Esta classe pode ser se um aluno foi aprovado ou reprovado, se uma pessoa possui uma doença ou não, dentre outras tantas possibilidades, sendo que nestes casos ou a previsão será uma ou outra. 

poderemos prever valores numéricos nos problemas de classificação, mas nestes casos, esta previsão sempre significará uma categoria. 

Ou seja, este número significa algo diferente do seu valor e ele poderia ser substituído por qualquer letra, palavra ou mesmo outro número, sem prejudicar o entendimento das previsões. 

Na prática, é muito comum encontrarmos os valores 0 e 1 representando classes. Por exemplo, o valor 0 pode indicar a reprovação de um aluno e o valor 1 a aprovação. Esta notação pode simplificar muito os códigos, uma vez que as linguagens de programação trabalham com números, e assim a classe já estará preparada para ser utilizada.

***-Regressão:***

 Os problemas de regressão existem quando precisamos prever um valor numérico específico. Este valor pode ser o preço de um produto, o peso ou altura de uma pessoa, a metragem de uma casa, e assim por diante. 

Nestas situações, o modelo poderá apresentar como resposta qualquer valor.

***-Clustering:*** 

ou agrupamento é uma técnica de análise de dados usada para dividir um conjunto de dados em grupos ou "clusters" com base em suas características. O objetivo do agrupamento é identificar padrões nos dados e agrupar amostras semelhantes em um mesmo cluster. 

Em contraste com a classificação, onde as Exceção de entrada já possuem uma classe pré-determinada, o agrupamento é uma técnica não supervisionada, ou seja, não há uma classe pré-definida para as Exceção. Em vez disso, o modelo de agrupamento é responsável por encontrar similaridades nos dados e agrupá-los em clusters. 

Existem várias técnicas de agrupamento, incluindo o k-means, o agrupamento hierárquico e o DBSCAN. O k-means é um dos algoritmos de agrupamento mais populares e é baseado em encontrar k clusters no conjunto de dados, onde k é o número de clusters definidos pelo usuário. O agrupamento hierárquico é outra técnica popular que divide o conjunto de dados em um conjunto de clusters aninhados. 

O agrupamento é usado em muitas aplicações de aprendizado de máquina, como análise de segmentação de clientes, análise de agrupamento de documentos e detecção de anomalias em dados. É uma técnica poderosa para explorar a estrutura de grandes conjuntos de dados e identificar padrões que podem não ser aparentes à primeira vista.

# ****Supervisionado:***

 é uma das abordagens mais comuns em machine learning, na qual um modelo é treinado para prever uma saída (ou "rótulo") para cada entrada (ou "característica") com base em exemplos rotulados de entrada e saída.

Em outras palavras, durante o treinamento,** o modelo é apresentado a um conjunto de dados que contém pares de entrada e saída correspondente.** Com base nesses exemplos, **o modelo aprende a mapear as entradas para as saídas correspondentes e pode, então, fazer previsão precisas para novos exemplos de entrada que nunca viu antes.**

Por exemplo, imagine que você queira construir um modelo de aprendizado de máquina que possa classificar imagens de animais em cães e gatos. Você forneceria ao modelo um conjunto de dados que contém imagens de animais rotulados como "cão" ou "gato" e, em seguida, treinaria o modelo para associar as características das imagens aos símbolos correspondentes. Depois de treinado, o modelo poderia ser usado para classificar novas imagens de animais em cães ou gatos com base nas características das imagens.

 **O objetivo é que o algoritmo aprenda a mapear as entradas para as saídas corretas, para que possa generalizar e prever corretamente as saídas para novos dados.**

 **Os exemplos de entrada e saída são divididos em um conjunto de treinamento e um conjunto de teste.** O conjunto de treinamento é usado para treinar o algoritmo, enquanto o conjunto de teste é usado para avaliar a precisão do algoritmo em novos dados.

O aprendizado supervisionado é comumente usado em problemas de **classificação,** onde o objetivo é prever uma classe ou categoria a que pertence a uma determinada entrada, bem como em problemas de **regressão**, onde o objetivo é prever um valor numérico.


# **Matriz de confusão,como calcular:**

Suponha que temos um problema de classificação binária em que estamos tentando prever se um paciente tem ou não uma determinada doença. Temos um conjunto de dados de teste com 100 amostras, das quais 50 são positivas para a doença e 50 são negativas. Suponha que um modelo de classificação foi treinado para realizar essa tarefa e, quando testado no conjunto de dados de teste, produziu as seguintes previsões:

Uma matriz de confusão é organizada em quatro quadrantes, como mostrado abaixo:

|             |  Previsto positivo | Previsto negativo |
|-------------|:------------------:|:------------------:|
| **Real positivo** |      Verdadeiro Positivo (VP)         |     Falso Negativo (FN)        |
| **Real negativo** |      Falso Positivo (FP)               |     Verdadeiro Negativo (VN)   |


Esses valores podem ser organizados em uma matriz de confusão:

|             |  Previsto positivo | Previsto negativo |
|-------------|:------------------:|:------------------:|
| **Real positivo** |         30         |         20         |
| **Real negativo** |         10         |         40         |

Podemos calcular as métricas de desempenho usando a matriz de confusão:

- Acurácia: (VP + VN) / (VP + VN + FP + FN) = (30 + 40) / 100 = 0,7 ou 70%
- Precisão: VP / (VP + FP) = 30 / (30 + 10) = 0,75 ou 75%
- Recall (sensibilidade): VP / (VP + FN) = 30 / (30 + 20) = 0,6 ou 60%
- F1-score: 2 * (precisão * recall) / (precisão + recall) = 2 * (0,75 * 0,6) / (0,75 + 0,6) = 0,67 ou 67%

Para entender a **diferença na acurácia em dados balanceados e desbalanceados**, suponha que o conjunto de dados de teste anterior tenha 90 amostras negativas para a doença e apenas 10 amostras positivas. Nesse caso, a matriz de confusão seria a seguinte:

|             |  Previsto positivo | Previsto negativo |
|-------------|:------------------:|:------------------:|
| **Real positivo** |         5          |         5          |
| **Real negativo** |         5          |         85         |

Nesse caso, a acurácia seria (VP + VN) / (VP + VN + FP + FN) = (5 + 85) / 100 = 0,9 ou 90%. Podemos ver que a acurácia aumentou, mas isso ocorreu principalmente porque o modelo está prevendo mais casos negativos corretamente (85) do que casos positivos corretamente (5). Nesse caso, é importante observar outras métricas, como a precisão e recall, para avaliar o desempenho do modelo de forma mais completa.

# ***Não supervisionado:**


Aprendizado não supervisionado é outra abordagem comum em Machine Learning, onde **o algoritmo é treinado em um conjunto de dados não rotulados, ou seja, dados em que as saídas desejadas não são conhecidas.** Ao contrário do aprendizado supervisionado, **o objetivo do aprendizado não supervisionado é encontrar estruturas e padrões nos dados de entrada, sem a necessidade de saídas rotuladas.**

O algoritmo é treinado para a**prender relações entre as diferentes variáveis ​​ou recursos do conjunto de dados, com o objetivo de agrupá-las em categorias ou clusters que seguem características semelhantes.** Por exemplo, em um conjunto de dados de imagens, o algoritmo pode identificar grupos de imagens com características semelhantes, como cores predominantes ou formas comuns.

Algumas técnicas comuns de aprendizado não supervisionadas incluem:

**Clustering**: onde o algoritmo agrupa as entradas em clusters com base em suas similaridades, sem ter informações prévias sobre as categorias.

Redução de dimensionalidade: onde o algoritmo reduz a complexidade do conjunto de dados, envolvendo as variáveis ​​em uma representação de menor dimensão, preservando ao mesmo tempo as informações importantes do conjunto de dados original.

Análise de componentes principais (PCA): onde o algoritmo busca as principais características dos dados, atendendo a dimensionalidade e destacando as correlações mais expressões.

O aprendizado não supervisionado é comumente usado em problemas de agrupamento, como segmentação de clientes ou detecção de anomalias em um conjunto de dados. Também pode ser usado para pré-processamento de dados antes de serem usados ​​em um algoritmo de aprendizado supervisionado.

# **Aprendizado por reforço:** 



Aprendizado por reforço é outra abordagem comum em Machine Learning, que é usado em situações em que **um agente deve tomar decisões em um ambiente dinâmico para maximizar uma recompensa.** Nesse caso,** o agente é colocado em um ambiente onde ele pode executar ações e, em seguida, recebe uma recompensa ou depósitos com base em suas ações.**

**O objetivo do aprendizado por segurança é treinar o agente para aprender a tomar as melhores ações possíveis em cada situação, de modo a maximizar a recompensa ao longo do tempo**. Para isso, o agente usa um algoritmo de aprendizado para aprender a associar ações com resultados esperados.

O processo de aprendizado por segurança **envolve três componentes principais: o agente, o ambiente e as recompensas**. O agente toma ações no ambiente e recebe uma recompensa ou descontos com base nessas ações. O ambiente é o contexto em que o agente está presente e pode ser qualquer coisa, desde um jogo até um sistema de controle industrial. A recompensa é uma medida que o agente usa para avaliar suas ações e pode ser positiva ou negativa.

O objetivo do agente **é encontrar a melhor estratégia para maximizar a recompensa ao longo do tempo, por meio de um processo de tentativa e erro**. O agente usa um algoritmo de aprendizado por reforço para aprender com suas experiências passadas e ajustar suas ações futuras.

Alguns algoritmos comuns de aprendizado por reforço incluem Q-Learning, SARSA e Policy Gradient Methods. O aprendizado por segurança é frequentemente usado em jogos de computador, robótica e sistemas de controle de processos industriais, onde o agente deve aprender a tomar decisões para otimizar o desempenho.

# **# Acurácia, precisão e recall:**

são indicadores comuns usados ​​para avaliar o desempenho de modelos de Machine Learning em **problemas de classificação.**

# **A acurácia**:
mede a proporção de exemplos classificados corretamente pelo modelo em relação ao número total de exemplos. A fórmula para calcular a acurácia é:

**Acurácia = (número de exemplos corretamente classificados) / (número total de exemplos)**

# **A precisão:** 

mede a proporção de exemplos positivos classificados corretamente em relação ao número total de exemplos classificados como positivos pelo modelo. A fórmula para calcular a precisão é:

**Precisão = (verdadeiros positivos) / (verdadeiros positivos + falsos positivos)**

# **O recall**

 mede a proporção de exemplos positivos classificados corretamente em relação ao número total de exemplos positivos presentes no conjunto de dados. A fórmula para calcular o recall é:

**Recall = (verdadeiros positivos) / (verdadeiros positivos + falsos negativos)**

Onde:

Verdadeiros positivos (TP): exemplos que foram corretamente classificados como positivos.

Falsos positivos (FP): exemplos que foram classificados negativamente como positivos.

Falsos negativos (FN): exemplos que foram classificados incorretamente como negativos.

Em geral, a precisão é importante quando o custo de um falso positivo é alto, enquanto o recall é importante quando o custo de um falso negativo é alto. Por exemplo, em um modelo de detecção de fraudes, um falso negativo pode custar muito mais do que um falso positivo.

É importante lembrar que essas métricas são úteis para avaliar modelos de classificação binária, mas para modelos de classificação multiclasse, é necessário usar outras métricas, como a matriz de confusão e a medida F1.

Além disso,**é importante avaliar o desempenho do modelo em um conjunto de dados de teste separado do conjunto de dados de treinamento, para garantir que o modelo esteja generalizando bem para novos dados e não esteja superajustando (overfitting) aos dados de treinamento.**

# **Minimizar o EQM (Erro Quadrático Médio)** 

EQM (Erro Quadrático Médio) é uma métrica usada para avaliar a precisão de um modelo de aprendizado de máquina em relação aos seus valores previstos em relação aos valores reais. 

**Quanto menor for o EQM, melhor será a qualidade do modelo de regressão**, pois isso significa que ele está fazendo previsões mais precisas e próximas aos valores reais.

O processo de minimização do EQM envolve o ajuste dos parâmetros do modelo para que ele seja capaz de fazer previsões mais precisas. Esse ajuste é geralmente feito usando algoritmos de otimização, como o Gradiente Descendente, que iterativamente ajustam os valores dos parâmetros para minimizar o EQM. O processo de treinamento é interrompido quando o valor do EQM não melhora mais significativamente, ou quando o modelo atinge uma precisão suficiente alta.

Em resumo,**minimizar o EQM é um objetivo comum em Machine Learning, especialmente em modelos de regressão, pois permite que o modelo faça prever mais precisas e se aproximar dos valores reais, aumentando sua capacidade de generalização e, portanto, sua capacidade de prever valores para dados novos e não vistos anteriormente**


O cálculo da EQM (Erro Quadrático Médio) envolve três etapas:

1. Calcule as previsões do modelo para todos os exemplos de treinamento:
- Para cada exemplo de treinamento, aplique a equação da regressão linear usando os pesos (coeficientes) aprendidos pelo modelo durante o treinamento;
- Isso resultará em uma previsão para o valor de y (variável dependente) para cada exemplo de treinamento.

2. Calcule a diferença entre as previsões do modelo e os valores reais:
- Para cada exemplo de treinamento, calcule a diferença entre a previsão do modelo e o valor real de y;
- Isso resultará em um valor de erro para cada exemplo de treinamento.

3. Calcule a média dos erros quadráticos:
- Para cada erro de cada exemplo de treinamento, calcule o erro ao quadrado (isto é, eleve o valor ao quadrado);
- Some todos os erros ao quadrado para todos os exemplos de treinamento;
- Divida a soma pelo número total de exemplos de treinamento (n).

A fórmula completa da EQM é:

EQM = (1/n) * Σ(yi - ȳ)^2

Onde:
- n é o número de exemplos de treinamento;
- yi é o valor real de y para o i-ésimo exemplo de treinamento;
- ȳ é a média dos valores reais de y em todo o conjunto de dados de treinamento;
- Σ é a soma sobre todos os exemplos de treinamento.

Por exemplo, suponha que você tenha um conjunto de dados de treinamento com n = 5 exemplos e a seguinte tabela de valores reais e previsões do modelo:

| Valor real (y) | Previsão do modelo |
|----------------|--------------------|
| 2              | 3                  |
| 5              | 4                  |
| 7              | 6                  |
| 8              | 8                  |
| 10             | 9                  |

Para calcular a EQM, primeiro calcule os erros para cada exemplo de treinamento:

| Valor real (y) | Previsão do modelo | Erro               |
|----------------|--------------------|--------------------|
| 2              | 3                  | (2 - 3)^2 = 1       |
| 5              | 4                  | (5 - 4)^2 = 1       |
| 7              | 6                  | (7 - 6)^2 = 1       |
| 8              | 8                  | (8 - 8)^2 = 0       |
| 10             | 9                  | (10 - 9)^2 = 1      |

Em seguida, some os erros ao quadrado:

EQM = (1/5) * (1 + 1 + 1 + 0 + 1) = 0,8

O resultado é a EQM média para os cinco exemplos de treinamento. Um valor mais baixo de EQM indica um modelo de regressão linear mais preciso.
